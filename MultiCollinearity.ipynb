```python
#importing library
import statsmodels.api as sm
import pandas as pd
```

```python
#reading data
df=pd.read_csv('data/salary.csv')
```

```python
df.head()
```

```python
#test for multiCollinearity by observing p value and std error and check rsquare to check how generalize it is
x=df.drop('Salary',axis=1)
y=df.Salary

x=sm.add_constant(x)
x.head()

model=sm.OLS(y,x).fit()
model.summary()
```
